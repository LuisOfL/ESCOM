## Limpieza de datos

### importar librerias

```python
import pandas as pd
```

```python
df = pd.read_csv("final.csv")
```

```python
df.head()
```

### Combinar cvs para tener la columna incidente

```python
df2 = pd.read_csv("911_clean.csv")
df["fecha_creacion"] = pd.to_datetime(df["fecha_creacion"])
df2["fecha_creacion"] = pd.to_datetime(df2["fecha_creacion"])

columnas_interes = ["fecha_creacion", "latitud", "longitud", "incidente_c4"]
df1_filtrado = df2[columnas_interes]

new_df = pd.merge(
    df, df1_filtrado, on=["fecha_creacion", "latitud", "longitud"], how="left"
)

new_df.head()
```

#### Convertir tipos de datos

```python
# latitud, longitud y las demas a float
new_df["latitud"] = new_df["latitud"].astype(float)

new_df["longitud"] = new_df["longitud"].astype(float)

new_df["temp"] = new_df["temp"].astype(float)

new_df["precip"] = new_df["precip"].astype(float)

new_df["humedad"] = new_df["humedad"].astype(float)

new_df["viento"] = new_df["viento"].astype(float)

new_df["radiacion"] = new_df["radiacion"].astype(float)

new_df["presion"] = new_df["presion"].astype(float)


# alcaldia_cierre y colonia cierre como str
new_df["alcaldia_cierre"] = new_df["alcaldia_cierre"].str.strip()
new_df["alcaldia_cierre"] = new_df["alcaldia_cierre"].astype(str)

new_df["colonia_cierre"] = new_df["colonia_cierre"].str.strip()
new_df["colonia_cierre"] = new_df["colonia_cierre"].astype(str)
```

```python
new_df.head()
```

### Verificar que las variables string se puedan convcertir a categoricas para usar para predecir

```python
new_df["alcaldia_cierre"].unique()
```

```python
new_df["colonia_cierre"].unique()
```

### Cambiar el nombre de las columnas

```python
new_df = new_df.rename(
    columns = {
        "fecha_creacion" : "fecha",
        "alcaldia_cierre" : "alcaldia",
        "colonia_cierre" : "colonia",
        "incidente_c4" : "incidente"
    }
    
)
```

### Exportar el csv

```python
new_df.to_csv("data.csv",index=False)
```
