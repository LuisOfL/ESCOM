## Funcion de Informacion mutua

Que tan relacionada esta una variable con otra

```python
X = list(df['Estudia'])
Y = list(df['Aprueba'])

def Marginals(X,Y):
    ProbaT = [0]*4
    for x in range(len(X)):
        if X[x] == 'Si':
            ProbaT[0] += 1
        if X[x] == 'No':
            ProbaT[1] += 1
        if Y[x] == 'Si':
            ProbaT[2] += 1
        if Y[x] == 'No':
            ProbaT[3] += 1
    return [x/len(X) for x in ProbaT]

def Contingence(X,Y):
    ProbaC = [0]*4
    for x in range(len(X)):
        if X[x] == 'Si' and Y[x] == 'Si':
            ProbaC[0] += 1
        if X[x] == 'Si' and Y[x] == 'No':
            ProbaC[1] += 1
        if X[x] == 'No' and Y[x] == 'No':
            ProbaC[2] += 1
        if X[x] == 'No' and Y[x] == 'Si':
            ProbaC[3] += 1
    return [x/len(X) for x in ProbaC]
    
def Test(t1,t2,Marginals,Contingence):
    p_x_y = 0
    p_x = 0
    p_y = 0
    if t1 == 'Si':
        p_x = Marginals[0]
        if t2 == 'Si':
            p_x_y = Contingence[0]
        if t2 == 'No':
            p_x_y = Contingence[1]
            
    if t1 == 'No':
        p_x = Marginals[1]
        if t2 == 'Si':
            p_x_y = Contingence[3]
        if t2 == 'No':
            p_x_y = Contingence[2]
            
    if t2 == 'Si':
        p_y = Marginals[2]
    if t2 == 'No':
        p_y = Marginals[3]
    if p_y == 0 or p_x == 0 or p_x_y == 0:
        return 0
    return p_x_y * math.log(p_x_y/(p_x*p_y))

def MutualInformation(X,Y):
    res = []
    res.append(Test('Si','Si',Marginals(X,Y),Contingence(X,Y)))
    res.append(Test('Si','No',Marginals(X,Y),Contingence(X,Y)))
    res.append(Test('No','No',Marginals(X,Y),Contingence(X,Y)))
    res.append(Test('No','Si',Marginals(X,Y),Contingence(X,Y)))
    return sum(res)
        
    
print('----------------ESTUDIA-----------------------')
print(Marginals(X,Y))
print(Contingence(X,Y))
print(MutueInformation(X,Y))
print('----------------DUERME BIEN--------------------')
X = list(df['DuermeBien'])
print(Marginals(X,Y))
print(Contingence(X,Y))
print(MutualInformation(X,Y))
```
