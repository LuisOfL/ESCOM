## Funcion de Informacion mutua

```r
X <- datos$Estudia
Y <- datos$Aprueba

MarginalS <- function(X,Y){
    probaM = c(0,0,0,0)
    for(x in 1:length(X)){
        if(X[x] == 'Si'){
            probaM[1] = probaM[1]+1
        }
        if(X[x] == 'No'){
            probaM[2] = probaM[2]+1
        }
        if(Y[x] == 'Si'){
            probaM[3] = probaM[3]+1
        }
        if(Y[x] == 'No'){
            probaM[4] = probaM[4]+1
        }
    }
    return(probaM/length(X))
}
Contingence <- function(X,Y){
     probaC = c(0,0,0,0)
    for(x in 1:length(X)){
        if(X[x] == 'Si' && Y[x] == 'Si'){
            probaC[1] <- probaC[1]+1
        }
        if(X[x] == 'Si' && Y[x] == 'No'){
            probaC[2] <- probaC[2]+1
        }
        if(X[x] == 'No' && Y[x] == 'No'){
            probaC[3] <- probaC[3]+1
        }
        if(X[x] == 'No' && Y[x] == 'Si'){
            probaC[4] <- probaC[4]+1
        }
    }
    return(probaC/length(X))
}
Test <- function(t1,t2,M,C){
    p_x_y <- 0
    p_y <- 0
    p_x <- 0
    if(t1 == 'Si'){
        p_x = M[1]
        if(t2 == 'Si'){
            p_x_y =  C[1]
        }
        if(t2 == 'No'){
            p_x_y =  C[2]
        }
    }
    if(t1 == 'No'){
        p_x = M[2]
        if(t2 == 'Si'){
            p_x_y =  C[4]
        }
        if(t2 == 'No'){
            p_x_y =  C[3]
        }
    }
    if(t2 == 'Si'){
        p_y = M[3]
    }
    if(t2 == 'No'){
        p_y = M[4]
    }
    if(p_y == 0 || p_x == 0 || p_x_y == 0){
        return(0)
    }
    return(p_x_y * log(p_x_y/(p_x*p_y)))
}

MutualInformation <- function(X,Y){
    res = c()
    res <- append(Test('Si','Si',MarginalS(X,Y),Contingence(X,Y)),res)
    res <- append(Test('Si','No',MarginalS(X,Y),Contingence(X,Y)),res)
    res <- append(Test('No','Si',MarginalS(X,Y),Contingence(X,Y)),res)
    res <- append(Test('No','No',MarginalS(X,Y),Contingence(X,Y)),res)
    return(sum(res))
}


print('--------------ESTUDIA--------------')
print(MarginalS(X,Y))
print(Contingence(X,Y))
print(MutualInformation(X,Y))
print('------------DUERME BIEN------------')
X <- datos$DuermeBien
print(MarginalS(X,Y))
print(Contingence(X,Y))
print(MutualInformation(X,Y))
```
